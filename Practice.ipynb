### Practice Functions Using One Parameter

```python
def greet(name):
    print("Hello",name)
greet('Awais')
greet("Raza")
greet("Waqas")
```

```python
def  my_city(city_name):
    print(f"I Live In {city_name}")
my_city("Islamabad")
my_city("Lahore")
```

```python
def food(favouritefood):
    print(f"I like {favouritefood}")
food("Biryani")
food("Pizza")

def age(myAge):
    print(f"I'm {myAge} years old.")
age(21)

def animal(animal_name):
    print(f"My Favourite Animal Is {animal_name}")
animal("Lion")
```
