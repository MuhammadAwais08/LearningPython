### Practice Functions Using One Parameter

```python
def greet(name):
    print("Hello",name)
greet('Awais')
greet("Raza")
greet("Waqas")
```

```python
def  my_city(city_name):
    print(f"I Live In {city_name}")
my_city("Islamabad")
my_city("Lahore")
```

```python
def food(favouritefood):
    print(f"I like {favouritefood}")
food("Biryani")
food("Pizza")

def age(myAge):
    print(f"I'm {myAge} years old.")
age(21)

def animal(animal_name):
    print(f"My Favourite Animal Is {animal_name}")
animal("Lion")
```

### Practice Functions Using Two Parameters

```python
def add(x,y):
    print(f"The Sum Of Two Numbers Is {x+y}")
add(10,5)
add(20,34)
add(234,867)
#Multilplication Of Three Numbers
def multi(x,y,z):
    print("The Multiplication Of Three Numbers Is =",x*y*z)
multi(1,2,3)
multi(34,43,33)
multi(19,27,36)
```

```python
def intro(name,age,purpose_of_Life):
    print(f"My Name Is : {name} \nMy Age Is: {age}\nMy Purpose Of Life Is {purpose_of_Life} ")
intro('Muhammad Awais Raza',21,'To Become A AI Engineer')
```

```python
def add(num1,num2):
    c=num1+num2
    return c
result=add(10,20)
print("The Sum Of Two Numbers Is =",result)

def square(num):
    return num*num
result=square(5)
print("The Square Of The Number Is =",result)

def multi(x,y):
    z=x*y
    return z
result=multi(10,23)
print("The Multiplication Of Two Numbers Is =",result)
result=multi(14,24)
print("The Multiplication Of Two Numbers Is =",result)
```

```python
def my_ful_name(first_name, Last_name):
    full_name=first_name+Last_name
    return full_name
result=my_ful_name("Muhammad Awais", "Raza (Future AI Engineer)")
print("My Full Name Is ",result)
```

```python
def even_num(num):
    if num%2==0:
        return True
    else:
        return False
result=even_num(9)
print("Is The Number Even? ",result)
```
