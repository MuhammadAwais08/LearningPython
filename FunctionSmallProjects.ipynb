### Mini Calculator Using Function

```python
def Calculator(a,b,Operation):
    if Operation=="Addition":
        c=a+b
        return c
    elif Operation=="Subtraction":
        c=a-b
        return c
    elif Operation=="Multilplication":
        c=a*b
        return c
    elif Operation=="Division":
        c=a/b
        return c
    else:
        return "Invalid Operation"
result=Calculator(10,6,"Division")
print("The Result Of The Operation Is: ",result)
result=Calculator(10,6,"Multilplication")
print("The Result Of The Operation Is: ",result)
result=Calculator(10,6,"Subtraction")
print("The Result Of The Operation Is: ",result)
result=Calculator(10,6,"Addition")
print("The Result Of The Operation Is: ",result)
```
